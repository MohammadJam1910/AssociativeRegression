```python
import pandas as pd
from scipy.io import arff
data=arff.loadarff('converted.arff')
df=pd.DataFrame(data[0])
transactions=df.apply(lambda row: list(row), axis=1).tolist()
print(transactions[:5])
```

```python
import pyfpgrowth

# Define a minimum support threshold
min_support = 2  # You can adjust this threshold

# Find frequent patterns
patterns = pyfpgrowth.find_frequent_patterns(transactions, min_support)
# Optionally, save the patterns to a CSV file
with open('frequent_patterns.csv', 'w') as f:
    for pattern, support in patterns.items():
        f.write(f'{pattern} #SUP: {support}\n')
```
